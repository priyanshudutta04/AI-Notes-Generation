## Importing Necessary Modules

```python
from transformers import pipeline as pipe
from google.colab import files
import io
import fitz
from fpdf import FPDF
import re
import PyPDF2
```

## Converting Pdf to text and breaking into chunks

```python
uploaded = files.upload()

pdf_filename = list(uploaded.keys())[0]
```

```python
def extract_text_from_pdf(save_path):
    pdf_document = fitz.open(save_path)
    text = ''
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        text += page.get_text() + "\n"
    return text


text = extract_text_from_pdf(pdf_filename)

print(text)
```

```python
word_count=len(text.split())
print(word_count)
```

```python
total_parts=round(word_count/500)

if total_parts==0:
  total_parts=1

print(total_parts)
```

```python
def split_text(text, n):

    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    cleaned_sentences = [sent.strip() for sent in sentences if sent.strip()]

    total_length = sum(len(sent) for sent in cleaned_sentences)
    part_length = total_length // n

    current_length = 0
    parts = []
    current_part = []


    for sentence in cleaned_sentences:
        sentence_length = len(sentence)
        if current_length + sentence_length > part_length and len(parts) < n-1:
            parts.append(' '.join(current_part))
            current_part = []
            current_length = 0
        current_part.append(sentence)
        current_length += sentence_length


    parts.append(' '.join(current_part))

    return parts



parts = split_text(text,total_parts)
```

## Summarization Text and Generating Title


```python
summarizer=pipe("summarization",model="t5-base",tokenizer="t5-base",framework="pt")
```

```python
question_pipeline = pipe("text2text-generation",model="google/flan-t5-base",framework="pt")
```

```python
generated_summary_arr = []

for i in parts:
  summary = summarizer(i)[0]['summary_text']
  generated_summary_arr.append(summary)
```

```python
gen_summary=" ".join(generated_summary_arr)
print(gen_summary)
```

```python
title=question_pipeline("Generate a title for the following abstract:"+
                        gen_summary,
                        max_length=8,
                        min_length=2,
                        do_sample=False,
                        length_penalty=0.3,
                        num_beams=2,
                        early_stopping=False)
```

```python
title_text=title[0]['generated_text']
print(title_text)
```

## Answering Questions

```python
qa_pipeline = pipe("question-answering",model="deepset/roberta-base-squad2",tokenizer="deepset/roberta-base-squad2",framework="pt")
```

```python
result = qa_pipeline(question="WHen was harrapa Civilization Created", context=text)
print(f"Question: WHen was harrapa Civilization Created")
print(f"Answer: {result['answer']}")
```

## Questions Generation

```python
generated_questions = []

for i in parts:
  question = question_pipeline(
      "Generate a question:"+i, max_length=50, num_return_sequences=1)[0]['generated_text'].strip()
  generated_questions.append(question)

for i in generated_questions:
    print(i)
```

## Formatting Text and generating Pdf

```python
def capitalize(text):
    sentences = re.split('(?<=[.!?]) +', text)
    sentences = [sentence.capitalize() for sentence in sentences]
    return ' '.join(sentences)

def remspace(text):
    cleaned_text = re.sub(r' \.', '.', text)
    return cleaned_text

def wordcap(text):
    words = text.split()
    capitalized_words = [word.capitalize() for word in words]
    return ' '.join(capitalized_words)
```

```python
title_text=wordcap(title_text)

rem=remspace(gen_summary)
summarized_text=capitalize(rem)

print(wordcap(title_text))
print(summarized_text)

idx=1
all_questions=[]
for i in generated_questions:
    print(f"Question {idx}: {i}")
    all_questions.append(f"Question {idx}: {i}")
    idx+=1
```

```python
def create_pdf(text, title, ques, filename):
    pdf = FPDF()

    pdf.add_page()

    pdf.set_font("Arial", size=20)

    pdf.cell(0, 10, title, align="C")
    pdf.ln(20)

    pdf.set_font("Arial", size=14)
    pdf.multi_cell(0, 10, text)
    pdf.ln(20)

    pdf.set_font("Arial", size=18)
    pdf.cell(0, 10, "Practice Questions", align="C")
    pdf.ln(15)

    pdf.set_font("Arial", size=14)
    for question in ques:
        pdf.set_font("Arial", size=14)
        pdf.multi_cell(0, 10, question)
        pdf.ln(2)

    pdf.output(filename)
```

```python
create_pdf(summarized_text,title_text,all_questions, f"{title_text}.pdf")
```
